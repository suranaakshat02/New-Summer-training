```python
#importing the files

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset

df = pd.read_csv("cc general.csv")
```

```python
## Data preprocessing

from sklearn.preprocessing import LabelEncoder,StandardScaler

enc = LabelEncoder()

df['CUST_ID'] = enc.fit_transform(df['CUST_ID'])

```

```python
df
```

```python
## Data scaling

sc = StandardScaler()
df.iloc[:,2:] = sc.fit_transform(df.iloc[:,2:])

```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
from sklearn.cluster import KMeans
wcss = []
clus = [i for i in range(1,11)]

for i in clus:
    km = KMeans(n_clusters = i,init = 'k-means++')
    km.fit(df.values)
    wcss.append(km.inertia_)
    
    
```

```python
plt.plot(clus,wcss)
plt.xlabel("Clusters")
plt.ylabel("WCSS")
```

```python
km = KMeans(n_clusters = 3,init = "k-means++")
km.fit(df)
```

```python
a = list(km.labels_).count(0)
b = list(km.labels_).count(1)
c = list(km.labels_).count(2)
```

```python
print("Cluster 1 count {}\n Cluster 2 count {} \n Cluster 3 count{}".format(a,b,c))
```
